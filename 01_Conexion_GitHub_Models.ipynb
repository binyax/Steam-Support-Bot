# 4. Implementación de la Conexión
En esta celda configuramos el cliente. Se incluye una lógica de verificación para asegurar que la conexión sea estable y reconozca las credenciales necesarias.

````python
from openai import OpenAI
import os

# Configuración del cliente OpenAI para GitHub Models
try:
    # Intentamos obtener las credenciales de las variables de entorno
    base_url = os.environ.get("GITHUB_BASE_URL", "https://models.inference.ai.azure.com")
    api_key = os.environ.get("GITHUB_TOKEN")

    if not api_key:
        raise ValueError("⚠️ GITHUB_TOKEN no encontrado. Revisa la configuración de tus variables de entorno.")

    client = OpenAI(
        base_url=base_url,
        api_key=api_key
    )
    
    print("✅ Cliente configurado exitosamente.")
    print(f"Base URL: {client.base_url}")
    # Preview seguro del token
    print(f"Token Validado: {api_key[:10]}...{api_key[-4:]}")

except Exception as e:
    print(f"❌ Error en la configuración: {e}")
````

# 5. Prueba de Concepto: Primera Consulta de Steam
Validaremos la conexión realizando una consulta simple sobre eventos de Steam. Esto confirma que el modelo tiene conocimiento base actualizado.

Utilizamos el system role para moldear la personalidad del asistente. Aquí es donde empezamos a aplicar el concepto de Prompt por Capas.

````python
try:
    response = client.chat.completions.create(
        messages=[
            {
                "role": "system", 
                "content": "Eres un experto en la plataforma Steam."
            },
            {
                "role": "user", 
                "content": "¿Cuándo suelen ser las rebajas de verano de Steam?"
            }
        ],
        model="gpt-4o",
        temperature=0.7,
        max_tokens=150
    )

    print("--- Respuesta del Modelo ---")
    print(response.choices[0].message.content)

except Exception as e:
    print(f"Error al realizar la consulta: {e}")
````
