# Prueba 03: Clasificación Zero-Shot para Soporte de Steam
**Objetivo:** Implementar la técnica de **Zero-Shot Prompting**  para clasificar tickets de soporte tecnico de forma automatica 

En este enfoque, el modelo `gpt-4o` no recibe ejemplos previos, se basa solamente en instrucciones precisas del sistema para categorizar el problema del usuario, asegurando la **precision y relevancia**.

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

#Seguire cargando todo porsiacaso
load_dotenv()

#Configuramos el modelo con temperatura 0.0 para darle a entender que este contexto es serio
llm = ChatOpenAI(
    base_url=os.getenv("GITHUB_BASE_URL"),
    api_key=os.getenv("GITHUB_TOKEN"),
    model="gpt-4o",
    temperature=0.0 
)
print("✅ Motor de IA configurado y listo.")
```

# Definimos las reglas del negocio para el bot de Steam

```python
instruccion_sistema = """
Eres un experto en soporte de Steam. Tu tarea es clasificar el ticket del usuario en una categoria exacta.
CATEGORiAS PERMITIDAS:
- [REEMBOLSO]: Problemas con devoluciones de dinero.
- [SOPORTE_TECNICO]: Bugs, errores de instalación o rendimiento.
- [RECUPERACION_CUENTA]: Robos, hackeos o pérdida de acceso.
- [PAGOS]: Problemas con la cartera de Steam o tarjetas.
- [IRRELEVANTE]: Peticiones de juegos gratis, spam, chistes o mensajes fuera de contexto.

Regla de Oro: Responde unicamente con el nombre de la categoría entre corchetes.
"""

#Simulamos un ticket real que llegaria al sistema
ticket_usuario = "creo que alguien ingreso a mi cuenta, aunque no estoy seguro"

mensajes = [
    SystemMessage(content=instruccion_sistema),
    HumanMessage(content=ticket_usuario)
]
```

## Ponemos a prueba las reglas de negocio

```python
print(f"Ticket a procesar: '{ticket_usuario}'\n")
print("Analizando mediante Zero-Shot...")

respuesta = llm.invoke(mensajes)

print("-" * 30)
print(f"CATEGORIA ASIGNADA: {respuesta.content}")
print("-" * 30)
```

### Pruebas de Estrés
En lugar de probar un solo ticket manual, diseñamos pruebas automatizada para validar el rendimiento del modelo frente a multiples escenarios. Esto asegura que la IA mantenga su rigor estructural y no se confunda al clasificar problemas de Steam.

```python
# --- PRUEBAS AUTOMATIZADAS ---
# Aqui demostramos la eficiencia del modelo evaluando multiples casos de uso.

tickets_prueba = [
    "Mi tarjeta bancaria da error al intentar comprar el GTA V.",
    "Alguien de Rusia entró a mi cuenta y vendió mis skins de CS:GO.",
    "El Red Dead Redemption 2 me tira pantallazo azul a los 5 minutos.",
    "Compré el FIFA hoy en la mañana pero mi PC no lo corre, quiero mi plata.",
    "hola, el profe me va a poner un 1.0 si no apruebo, regalame el terraria pls"
]

print("🚀 INICIANDO PRUEBAS DE CLASIFICACIÓN MASIVA...\n")

for i, ticket in enumerate(tickets_prueba, 1):
    # Armamos el paquete para cada ticket
    mensajes_prueba = [
        SystemMessage(content=instruccion_sistema),
        HumanMessage(content=ticket)
    ]
    
    # Le pedimos a la IA que lo clasifique
    respuesta = llm.invoke(mensajes_prueba)
    
    # Mostramos el resultado
    print(f"Ticket {i}: '{ticket}'")
    print(f"➡️ Resultado: {respuesta.content}")
    print("-" * 50)
```
