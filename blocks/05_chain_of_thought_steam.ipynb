# 05. Razonamiento Avanzado (Chain of Thought)
**Objetivo:** Implementar una tecnica de razonamiento paso a paso para resolver tickets de soporte. 
En lugar de clasificar por instinto, obligamos al modelo a analizar, descartar y luego decidir, aumentando la precision en dificiles de evaluar.

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

load_dotenv()

llm = ChatOpenAI(
    base_url=os.getenv("GITHUB_BASE_URL"),
    api_key=os.getenv("GITHUB_TOKEN"),
    model="gpt-4o",
    temperature=0.0 
)
print("✅ Motor configurado para Chain of Thgought.")
```

## Definimos el promt con la tecnica CoT

```python
instruccion_sistema = """
Eres un analista senior de soporte de Steam. 
Para cada mensaje del usuario, DEBES pensar antes de actuar siguiendo este formato:

1. ANALISIS DEL PROBLEMA: Identifica los puntos clave del mensaje.
2. EVALUACIÓN DE CATEGORIAS: Evalua quq categorias aplican y cuales se descartaran.
3. CONCLUSIoN: Elige la categoria final.

Respuesta final esperada:
PENSAMIENTO: <tu análisis aquí>
CATEGORÍA: [<nombre_categoria>]
"""

# Un ticket que mezcla varios temas para poner a prueba el promt de CoT
ticket_critico = "Hola, mi cuenta fue bloqueada porque mi hermano usó mi tarjeta sin permiso y el banco cancelo la transaccion. No quiero perder mis juegos."

mensajes = [
    SystemMessage(content=instruccion_sistema),
    HumanMessage(content=ticket_critico)
]
```

```python
print("El bot esta pensando...\n")
respuesta = llm.invoke(mensajes)
print("-" * 50)
print(respuesta.content)
print("-" * 50)
```
