# 2.1. LangChain Streaming
## 1. Importación de Bibliotecas y Dependencias
Antes de iniciar, cargamos los módulos necesarios para la conectividad con la API, el manejo de mensajes de LangChain y el control del flujo de tiempo para el streaming.

```python
import os
import time
import sys
from dotenv import load_dotenv

# Componentes de LangChain
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

# Carga de variables de entorno (.env)
load_dotenv()

print("✅ Bibliotecas importadas y entorno cargado correctamente.")
```

## 2. ¿Qué es el Streaming?
El Streaming es la capacidad de recibir la respuesta del modelo token por token (palabra por palabra) en tiempo real, en lugar de esperar a que el servidor genere el bloque de texto completo.

Cuando el bot genera guías técnicas largas, el streaming evita que el usuario piense que la aplicación se "congeló".

## 3. Configuración del Motor de IA ⚙️
Habilitamos el parámetro streaming=True para que el modelo nos entregue la información en fragmentos.

```python
# Configuración del modelo con streaming habilitado
try:

    llm = ChatOpenAI(
        base_url=os.getenv("GITHUB_BASE_URL"),
        api_key=os.getenv("GITHUB_TOKEN"),
        model="gpt-4o",
        streaming=True,  # Es MUY necesario que streaming esté habilitado con True
        temperature=0.7,
        max_tokens=800
    )
    
    print(f"✅ Modelo {llm.model_name} configurado con streaming: {llm.streaming}")
    
except Exception as e:
    print(f"❌ Error en configuración: {e}. Verifica tu GITHUB_TOKEN.")
```

# 4. Streaming Básico

Para probar vamos a definir nuevamente el hardware del usuario

```python
USER_HARDWARE = {
    "device": "Asus Vivobook 14 X1404",
    "cpu": "Intel Core i5-1235U",
    "ram": "8GB DDR4",
    "os": "Windows 11"
}

print(f"✅ Hardware de prueba registrado: {USER_HARDWARE['device']}")
```

```python
def streaming_diagnostico_steam():
    prompt = f"Explícame por qué 'Grand Theft Auto V' podría tener tirones (stuttering) en mi {USER_HARDWARE} y cómo solucionarlo."
    
    print("=== 🛠️ INICIANDO DIAGNÓSTICO EN TIEMPO REAL ===")
    print("-" * 50)
    
    try:
        # El bucle captura cada chunk conforme llega del servidor de GitHub
        for chunk in llm.stream([HumanMessage(content=prompt)]):
            print(chunk.content, end="", flush=True)
            time.sleep(0.01) # Pequeña pausa para simular escritura humana
            
        print("\n" + "-" * 50)
        print("✓ Diagnóstico completado.")
        
    except Exception as e:
        print(f"❌ Error en streaming: {e}")

# Ejecutar prueba básica
streaming_diagnostico_steam()
```

## 5. Comparación Técnica: Streaming vs. No-Streaming

```python
def comparar_experiencia_soporte():
    # Modelo estándar (sin streaming)
    llm_no_stream = ChatOpenAI(
        base_url=os.getenv("GITHUB_BASE_URL"),
        api_key=os.getenv("GITHUB_TOKEN"),
        model="gpt-4o",
        streaming=False
    )
    
    pregunta = f"¿Qué juegos con reseñas mayormente positivas me recomiendas que aprovechen bien mi {USER_HARDWARE}? Nombra solo 3 juegos y una breve explicacion"

    print("=== COMPARACIÓN: SOPORTE ESTÁTICO vs. REACTIVO ===\n")
    
    # 1. Sin streaming
    print("1. MODO ESTÁTICO:")
    start_time = time.time()
    try:
        response = llm_no_stream.invoke([HumanMessage(content=pregunta)])
        end_time = time.time()
        print(f"⏱️ Tiempo de espera 'muerto': {end_time - start_time:.2f} segundos")
        print(f"Respuesta parcial: {response.content[:100]}...")
    except Exception as e: print(f"Error: {e}")

    print("\n" + "="*60 + "\n")
    
    # 2. Con streaming
    print("2. MODO STREAMING:")
    start_time = time.time()
    print("🤖 El soporte técnico está escribiendo: ", end="")
    try:
        for chunk in llm.stream([HumanMessage(content=pregunta)]):
            print(chunk.content, end="", flush=True)
            time.sleep(0.02)
        end_time = time.time()
        print(f"\n\n✅ Generación finalizada en {end_time - start_time:.2f} segundos")
    except Exception as e: print(f"Error: {e}")

# Ejecutar comparación
comparar_experiencia_soporte()
```

# 6. Chatbot Interactivo de Steam con Streaming
Implementamos el bucle final donde el usuario puede conversar con el soporte técnico.

```python
# --- CELDA DE CÓDIGO: CHATBOT DE SOPORTE INTERACTIVO ---

def chatbot_streaming_steam():
    print("=== 🛠️ SOPORTE TÉCNICO STEAM (MODO STREAMING) ===")
    print("Instrucciones: Escribe tu duda técnica. Escribe 'salir' para finalizar.\n")
    
    system_prompt = f"""Eres un experto en soporte técnico de Steam. 
    Tu objetivo es ayudar al usuario con su hardware específico: {USER_HARDWARE}.
    Debes ser técnico, preciso y honesto sobre el rendimiento de los juegos. 
    Responde siempre de manera educativa y amable."""
    
    while True:
        try:
            # 2. Captura de la entrada del usuario
            user_input = input("\n🧑 Usuario: ")
            
            # Condición de salida
            if user_input.lower() in ['salir', 'exit', 'quit', 'chao', 'bye']:
                print("\n👋 Soporte Steam: ¡Gracias por contactarnos! Que disfrutes tus juegos.")
                break
                
            if not user_input.strip():
                continue
                
            print("\n🤖 Asistente Steam: ", end="", flush=True)
            
            # 3. Preparación de mensajes para LangChain
            from langchain_core.messages import SystemMessage, HumanMessage
            
            messages = [
                SystemMessage(content=system_prompt),
                HumanMessage(content=user_input)
            ]
            
            # 4. Generación con streaming y guardado de la respuesta completa
            full_response = ""
            
            # Iteramos sobre los fragmentos (chunks) que envía la IA
            for chunk in llm.stream(messages):
                content = chunk.content
                print(content, end="", flush=True) # Imprime en la misma línea
                full_response += content
                time.sleep(0.01) # Simulación de fluidez visual
                
            print()  # Salto de línea al finalizar la respuesta
            
        except KeyboardInterrupt:
            print("\n\n⏸️ Sesión terminada por el usuario.")
            break
        except Exception as e:
            print(f"\n❌ Ocurrió un error inesperado: {e}")

# --- EJECUCIÓN ---
chatbot_streaming_steam()
```
