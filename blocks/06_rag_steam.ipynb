# 06. Arquitectura RAG (Retrieval-Augmented Generation)
**Objetivo:** Lograr una integración del modelo base (GPT-4o) con datos externos estructurados (`reglas_steam.txt`). Se implementa la cadena completa: Extracción (TextLoader), Fragmentación (CharacterTextSplitter), Vectorización (OpenAIEmbeddings) y Recuperación (VectorStore).

**Las 4 Fases de este módulo:**
1. **Configurar Motores:** LLM (para hablar) y Embeddings (para traducir texto a números).
2. **Ingesta y Fragmentación:** Leer el `.txt` y cortarlo en pedacitos (Chunks).
3. **Vectorización:** Guardar esos pedacitos matemáticamente en una base de datos.
4. **Recuperación (Retrieval):** Buscar la regla exacta y responder al ticket.

## Configuracion y motor

Aqui cargamos nuestras credenciales, el modelo de chat y el modelo de embeddings (el que convierte texto en numeros).

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI, OpenAIEmbeddings

load_dotenv()

llm = ChatOpenAI(
    base_url=os.getenv("GITHUB_BASE_URL"),
    api_key=os.getenv("GITHUB_TOKEN"),
    model="gpt-4o",
    temperature=0.0
)

# Motor de embeddings
embeddings = OpenAIEmbeddings( #
    base_url=os.getenv("GITHUB_BASE_URL"),
    api_key=os.getenv("GITHUB_TOKEN"),
    model="text-embedding-3-small" #El modelo este se supone que es uno pequeno, rapido y barato disenado especificamente para esta tarea
)

print("✅ Motores de Chat y Embeddings configurados.")
```
